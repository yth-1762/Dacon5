- ID: 고유 식별자, 각 행을 고유하게 식별하는 데 사용됩니다.
- Age: 나이, 개인의 나이를 나타냅니다.
- Gender: 성별, 개인의 성별을 나타냅니다 ('M'은 남성, 'F'는 여성).
- Education_Status: 교육 수준, 개인의 최종 교육 수준을 나타냅니다.
- Employment_Status: 고용 상태, 개인의 현재 고용 상태를 나타냅니다 (예: 전일제, 비고용 등).
- Working_Week (Yearly): 연간 근로 주, 1년 동안 근무한 주 수를 나타냅니다.
- Industry_Status: 산업 상태, 개인이 근무하는 산업 분야를 나타냅니다.
- Occupation_Status: 직업 상태, 개인의 현재 직업 유형을 나타냅니다.
- Race: 인종, 개인의 인종을 나타냅니다.
- Hispanic_Origin: 히스패닉 여부, 개인이 히스패닉인지 여부를 나타냅니다.
- Martial_Status: 결혼 상태, 개인의 결혼 상태를 나타냅니다.
- Household_Status: 가구 상태, 개인의 가구 내 위치를 나타냅니다.
- Household_Summary: 가구 요약, 가구 내에서 개인의 역할을 요약해서 나타냅니다.
- Citizenship: 시민권 상태, 개인의 시민권 상태를 나타냅니다.
- Birth_Country: 출생국, 개인이 태어난 국가를 나타냅니다.
- Birth_Country (Father): 부친의 출생국, 개인의 아버지가 태어난 국가를 나타냅니다.
- Birth_Country (Mother): 모친의 출생국, 개인의 어머니가 태어난 국가를 나타냅니다.
- Tax_Status: 세금 상태, 개인의 세금 신고 상태를 나타냅니다.
- Gains: 이득, 개인이 투자 등으로 얻은 소득을 나타냅니다.
- Losses: 손실, 개인이 투자 등으로 발생한 손실을 나타냅니다.
- Dividends: 배당금, 개인이 배당금으로 받은 소득을 나타냅니다.
- Income_Status: 소득 상태, 개인의 소득 수준을 나타내는 범주형 변수입니다.
- Income: 소득, 개인의 총 소득을 나타냅니다.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 모든 범주형 변수 선택
categorical_columns = train.select_dtypes(include=['object']).columns.tolist()
categorical_columns.remove('ID')  # ID 제외

# 범주형 변수의 분포 확인
fig, axes = plt.subplots(len(categorical_columns), 1, figsize=(10, 5*len(categorical_columns)))
for i, col in enumerate(categorical_columns):
    sns.countplot(y=col, data=train, ax=axes[i], order=train[col].value_counts().index)
    axes[i].set_title(f'Distribution of {col}')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


# seaborn 스타일 설정
sns.set(style="whitegrid")

# 수치형 변수 선택
numeric_columns = train.select_dtypes(include=['int64', 'float64']).columns.tolist()

# 수치형 변수의 분포 확인
fig, axes = plt.subplots(len(numeric_columns), 1, figsize=(8, 5 * len(numeric_columns)))
for i, col in enumerate(numeric_columns):
    sns.histplot(train[col], bins=30, kde=True, ax=axes[i])
    axes[i].set_title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```

```python
# 변수들 간의 상관관계 확인
corr = train[numeric_columns].corr()

# 히트맵으로 상관관계 시각화
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numeric Features')
plt.show()
```

```python
# 교육 수준별 평균 소득 계산
avg_income_by_education = train.groupby('Education_Status')['Income'].mean().sort_values()

# 교육 수준별 평균 소득 시각화
plt.figure(figsize=(10, 8))
sns.barplot(x=avg_income_by_education.values, y=avg_income_by_education.index, palette="coolwarm")
plt.title('Average Income by Education Status')
plt.xlabel('Average Income')
plt.ylabel('Education Status')
plt.show()
```

```python
# 결혼 상태별 평균 소득 계산
avg_income_by_marital_status = train.groupby('Martial_Status')['Income'].mean().sort_values()

# 결혼 상태별 평균 소득 시각화
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_income_by_marital_status.values, y=avg_income_by_marital_status.index, palette="viridis")
plt.title('Average Income by Marital Status')
plt.xlabel('Average Income')
plt.ylabel('Marital Status')
plt.show()
```

```python
# 산업 상태(Industry_Status)별 평균 소득 계산
avg_income_by_industry = train.groupby('Industry_Status')['Income'].mean().sort_values()

# 산업 상태별 평균 소득 시각화
plt.figure(figsize=(10, 12))
sns.barplot(x=avg_income_by_industry.values, y=avg_income_by_industry.index, palette="rocket")
plt.title('Average Income by Industry Status')
plt.xlabel('Average Income')
plt.ylabel('Industry Status')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 데이터 불러오기 (data는 여러분의 DataFrame 이름으로 대체하세요)

# 인종별 평균 소득 계산
race_avg_income = train.groupby('Race')['Income'].mean().sort_values()

# 인종별 평균 소득 시각화
plt.figure(figsize=(10, 6))
sns.barplot(x=race_avg_income.index, y=race_avg_income.values, palette="coolwarm")
plt.title('Average Income by Race')
plt.xlabel('Race')
plt.ylabel('Average Income')
plt.xticks(rotation=45)
plt.show()

# 직업 상태별 평균 소득 계산
occupation_avg_income = train.groupby('Occupation_Status')['Income'].mean().sort_values()

# 직업 상태별 평균 소득 시각화
plt.figure(figsize=(12, 8))
sns.barplot(x=occupation_avg_income.index, y=occupation_avg_income.values, palette="coolwarm")
plt.title('Average Income by Occupation Status')
plt.xlabel('Occupation Status')
plt.ylabel('Average Income')
plt.xticks(rotation=45)
plt.show()
```
